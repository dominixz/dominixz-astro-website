```python
import itertools
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Diff input and control data
start_forecast = '2024-10-01'
month_forecast = 15
ze_trend, ze_seasonal = (0, 1, 1), (2, 1, 0, 3)
wr_trend, wr_seasonal = (2, 1, 2), (1, 0, 0, 3)
res_trend, res_seasonal = (0, 1, 1), (0, 1, 0, 12)
mor_trend, mor_seasonal = (0, 1, 1), (2, 0, 0, 12)
resp_trend, resp_seasonal = (0, 1, 0), (0, 0, 0, 3)
oth_trend, oth_seasonal = (0, 1, 1), (1, 0, 0, 3)

# ZE
# 36.27021691212444
# 8.0
# (0, 1, 1)
# (0, 1, 1)
# (2, 1, 0, 3)

# WR
# 2.1267950889296205
# -1.1314692315225034
# (0, 1, 1)
# (2, 1, 2)
# (1, 0, 0, 3)

# RES
# Best ARIMA AIC: 56.96273745743193
# Best SARIMA AIC: 43.28072120026411
# Best SARIMAX AIC: 45.017746253188086 Sep'21 สะท้อนเดือน Jan'22 > Take log ก่อน
# Best SARIMAX AIC: 45.149714559771304 Oct'21 สะท้อนเดือน Jan'22
# Best SARIMAX AIC: 44.9278020397087 Nov'21 สะท้อนเดือน Jan'22 > Take log ก่อน
# Best SARIMAX AIC: 44.33368473384051 Dec'21 สะท้อนเดือน Jan'22 > Take log ก่อน
# Best SARIMAX AIC: 42.55409969095192 Jan'22 สะท้อน Jan'22 > Take log ก่อน
# Best SARIMAX AIC: 42.42034727870426
# Best ARIMA parameters: (0, 1, 1)
# Best SARIMA parameters: (0, 1, 1) with seasonal order (0, 1, 0, 12)
# Best SARIMAX parameters: (0, 1, 1) with seasonal order (0, 1, 0, 12)

# MOR
# Best ARIMA AIC: -7.370536855312004
# Best SARIMA AIC: -12.428331469077246
# Best SARIMAX AIC: -11.373674852978464 Sep'21 สะท้อนเดือน Jan'22 > Take log ก่อน
# Best SARIMAX AIC: -12.609579641303917 Oct'21 สะท้อนเดือน Jan'22 > Take log ก่อน
# Best SARIMAX AIC: -10.432328408080416 Nov'21 สะท้อนเดือน Jan'22 > Take log ก่อน
# Best SARIMAX AIC: -10.665451935892584 Dec'21 สะท้อนเดือน Jan'22 > Take log ก่อน
# Best SARIMAX AIC: -10.76013193264659 Jan'21 สะท้อนเดือน Jan'22 > Take log ก่อน
# Best ARIMA parameters: (0, 1, 1)
# Best SARIMA parameters: (0, 1, 1) with seasonal order (2, 0, 0, 12)

# RESP
# -49.960087187017024
# -49.960087187017024
# (0, 1, 0)
# (0, 1, 0)
# (0, 0, 0, 3)

# OTHER
# 21.640256983543658
# 20.417823771736053
# (0, 1, 1)
# (0, 1, 1)
# (1, 0, 0, 3)

# Adjust the start forecast date
start_forecast_adj = pd.to_datetime(start_forecast, format="%Y-%m-%d") - pd.DateOffset(days=1)

# Generate future dates for the forecast period
future_dates = pd.date_range(start=start_forecast_adj, periods=month_forecast, freq='M') + pd.DateOffset(days=1)
print(future_dates)

# Data
data = pd.read_excel('/Users/thirangkoon/Downloads/Revenue by product.xlsx')
service_focus = ['ZE', 'WR', 'RES', 'MOR', 'RESP', 'OTHER']
df = pd.DataFrame(data)
df.set_index('Month', inplace=True)
df.drop(df.index[df.index >= pd.to_datetime(start_forecast)], inplace=True)
display(df.tail(3))  # Display the first few rows of the DataFrame

data_output = pd.DataFrame()

# SARIMA Model
for x in service_focus:
    df_service = df[x]

    result = seasonal_decompose(df_service, model='additive', period=12)
    result.plot()
    plt.show()

    if x == 'ZE':
        trend = ze_trend
        seasonal = ze_seasonal
    elif x == 'WR':
        trend = wr_trend
        seasonal = wr_seasonal
    elif x == 'RES':
        trend = res_trend
        seasonal = res_seasonal
    elif x == 'MOR':
        trend = mor_trend
        seasonal = mor_seasonal
    elif x == 'RESP':
        trend = resp_trend
        seasonal = resp_seasonal
    else:
        trend = oth_trend
        seasonal = oth_seasonal
    
    # Set up plot
    fig, axes = plt.subplots(3, 2, figsize=(21, 12))

    # Plot original data
    df_service.plot(title='Original Time Series Data of {}'.format(x), ax=axes[0, 0])

    # Prepare new data: first differenced
    df_diff = df_service.diff().dropna()

    # Plot first differenced data
    df_diff.plot(title='First Differenced Time Series Data of {}'.format(x), ax=axes[0, 1])

    # Check original series stationarity using ADF test
    result_1 = adfuller(df_service.dropna())
    print(f'Service: {x}')
    print('ADF Statistic (Original):', result_1[0])
    print('p-value (Original):', result_1[1])
    if result_1[1] < 0.05:
        print("The original series is stationary.")
    else:
        print("The original series is non-stationary.")
    print('\n')

    # Check first differenced series stationarity using ADF test
    result_2 = adfuller(df_diff)
    print('ADF Statistic (First Diff):', result_2[0])
    print('p-value (First Diff):', result_2[1])
    if result_2[1] < 0.05:
        print("The differenced series is stationary.")
    else:
        print("The differenced series is non-stationary.")
    print('\n')

    # ACF and PACF plots
    plot_acf(df_diff, ax=axes[1, 0])
    axes[1, 0].set_title('Autocorrelation Function (ACF)')

    plot_pacf(df_diff, ax=axes[1, 1])
    axes[1, 1].set_title('Partial Autocorrelation Function (PACF)')

    # Fitting SARIMAX model
    model = SARIMAX(df_service, order=trend, seasonal_order=seasonal)
    model_fit = model.fit(disp=False)
    # Print model summary
    print(f"Model Summary for {x}:")
    print(model_fit.summary())
    print('\n')

    # Forecast for the next ?? months
    forecast = model_fit.get_forecast(steps=month_forecast)
    forecast_values = forecast.predicted_mean
    forecast_conf_int = forecast.conf_int()

    # Extract the last data point of the specific service
    last_data_point = pd.Series([df_service.iloc[-1]], index=[df_service.index[-1]])

    # Concatenate the last data point with the forecast values
    extended_forecast = pd.concat([last_data_point, forecast_values])

    # Histogram of residuals
    residuals = model_fit.resid
    axes[2, 0].hist(residuals, bins=30)
    axes[2, 0].set_title('Residuals Histogram')

    # Backtest
    pred = model_fit.predict()
    df_service = pd.DataFrame(df_service)
    df_service['predict'] = pred
    df_service['predict'].iloc[0] = df_service.iloc[0, 0]

    # Plotting forecast
    axes[2, 1].plot(df_service[x], label='Original', color='blue')
    axes[2, 1].plot(extended_forecast, label='Forecast', color='red')
    axes[2, 1].plot(df_service['predict'], label = 'Backtest', color = 'firebrick', linestyle = '--')
    axes[2, 1].fill_between(future_dates, forecast_conf_int.iloc[:, 0], forecast_conf_int.iloc[:, 1], color='pink', alpha=0.3)
    axes[2, 1].set_ylim(bottom=0)
    axes[2, 1].set_title(f'Forecast for {x}')
    axes[2, 1].set_xlabel('Date')
    axes[2, 1].set_ylabel('Value')
    axes[2, 1].legend()
    axes[2, 1].grid(True)
    plt.show()

    plt.tight_layout()

    # Collect forecasted values into a DataFrame
    forecast_df = pd.DataFrame({'Forecast': forecast_values}, index=future_dates)
    forecast_df.index.name = 'Date'
    forecast_df.columns = [x]  # Rename column to the service name
    data_output = pd.concat([data_output, forecast_df], axis=1)

# Display the consolidated forecast DataFrame
display(data_output)

# To excel
data_output.to_excel('prediction.xlsx')
```

```python
# Focus on service
service = 'MOR'

# Define ranges for non-seasonal and seasonal parameters
p = q = range(0, 3)
d = 1
P = Q = range(0, 3)
D = range(0, 2)
seasonal = [3, 12]

# Generate combinations of parameters
pq = list(itertools.product(p, q))
pdq = [(a[0], d, a[1]) for a in pq]
PDQS = list(itertools.product(P, D, Q, seasonal))

# Track the best models
best_arima_aic = float('inf')
best_sarima_aic = float('inf')
best_arima_pdq = None
best_sarima_pdq = None
best_sarima_PDQS = None

# Grid search for ARIMA (non-seasonal)
print("Searching for the best ARIMA model...")
for trends in pdq:
    try:
        # Fit ARIMA model (non-seasonal)
        arima_model = SARIMAX(df[service], order=trends,
                              enforce_stationarity=True, enforce_invertibility=False)
        arima_model = arima_model.fit(disp=False)
        
        # Update the best ARIMA model if a lower AIC is found
        if arima_model.aic < best_arima_aic:
            best_arima_aic = arima_model.aic
            best_arima_pdq = trends
        
        print(f"ARIMA{trends} - AIC: {arima_model.aic}")

    except Exception as e:
        print(f"An error occurred in ARIMA: {e}")
        continue

# Grid search for SARIMA (seasonal)
print("Searching for the best SARIMA model...")
for trends in pdq:
    for seasons in PDQS:
        try:
            # Fit SARIMA model (with seasonal components)
            sarima_model = SARIMAX(df[service], order=trends, seasonal_order=seasons,
                                   enforce_stationarity=True, enforce_invertibility=False)
            sarima_model = sarima_model.fit(disp=False)
            
            # Update the best SARIMA model if a lower AIC is found
            if sarima_model.aic < best_sarima_aic:
                best_sarima_aic = sarima_model.aic
                best_sarima_pdq = trends
                best_sarima_PDQS = seasons
        
            print(f"SARIMA{trends}x{seasons} - AIC: {sarima_model.aic}")

        except Exception as e:
            print(f"An error occurred in SARIMA: {e}")
            continue

# Display the results
print(f"\nBest ARIMA model: {best_arima_pdq}, AIC: {best_arima_aic}")
print(f"Best SARIMA model: {best_sarima_pdq} with seasonal order {best_sarima_PDQS}, AIC: {best_sarima_aic}")

# Compare ARIMA vs SARIMA
if best_arima_aic < best_sarima_aic:
    print(f"\nARIMA is better with AIC: {best_arima_aic}")
else:
    print(f"\nSARIMA is better with AIC: {best_sarima_aic}")

# Track the best models
print(best_arima_aic)
print(best_sarima_aic)
print(best_arima_pdq)
print(best_sarima_pdq)
print(best_sarima_PDQS)
```
